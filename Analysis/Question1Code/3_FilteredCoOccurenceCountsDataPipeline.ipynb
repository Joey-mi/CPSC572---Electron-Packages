# Overview

Take the total co-occurrences, sort them, and filter them

```python
# Link Google Drive data to this colab session
from google.colab import drive
drive.mount('/content/drive')

# Bring in important imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Create a dataframe of the total co-occurences
totalDF = pd.read_csv("/content/drive/MyDrive/CPSC 572 Project/Data/coOccurenceCounts.csv", sep="\t", dtype={'Dependency Pair': str, 'Co-Occurences': 'Int64'})
display(totalDF)
```

```python
# Aggregate similar rows
d = {'Co-Occurences': 'sum'}
df_new = totalDF.groupby('Dependency Pair', as_index=False).aggregate(d).reindex(columns=totalDF.columns)

# Sort desc by co-occurrences
sortedDF = df_new.sort_values(['Co-Occurences'], ascending=[False])
display(sortedDF)
```

```python
#Find dict...
dictWhere = totalDF[(totalDF['Dependency Pair'].str.contains("dict") == True)]
display(dictWhere)
```

```python
# Save as a CSV
import os

filename = "coOccurenceCountsAggregated.csv"

sortedDF.to_csv('/content/drive/MyDrive/CPSC 572 Project/Data/' + filename, sep='\t', index=False)
if os.path.isfile('/content/drive/MyDrive/CPSC 572 Project/Data/' + filename):
  print(f"Successfully uploaded {filename} to /content/drive/MyDrive/CPSC 572 Project/Data/")
else:
  print(f"ERROR: Download unsuccessful for {filename} to /content/drive/MyDrive/CPSC 572 Project/Data/")
```

```python
# List words to filter out
common1 = "react"
common2 = "electron"
common3 = "vue"
common4 = "angular"
common5 = "testing"
occurrence = 1 # 1 to filter common words out completely, 2 to filter rows where they are paired with each other, 3 for no filter

attemptFilter = sortedDF[(sortedDF['Dependency Pair'].str.count(common1) < occurrence)
& (sortedDF['Dependency Pair'].str.count(common2) < occurrence)
& (sortedDF['Dependency Pair'].str.count(common3) < occurrence)
& (sortedDF['Dependency Pair'].str.count(common4) < occurrence)
& (sortedDF['Dependency Pair'].str.count(common5) < occurrence)]
display(attemptFilter.head(25))
```

```python
# Save as a CSV
import os

filename = "coOccurenceCountsFiltered.csv"

attemptFilter.to_csv('/content/drive/MyDrive/CPSC 572 Project/Data/' + filename, sep='\t', index=False)
if os.path.isfile('/content/drive/MyDrive/CPSC 572 Project/Data/' + filename):
  print(f"Successfully uploaded {filename} to /content/drive/MyDrive/CPSC 572 Project/Data/")
else:
  print(f"ERROR: Download unsuccessful for {filename} to /content/drive/MyDrive/CPSC 572 Project/Data/")
```
