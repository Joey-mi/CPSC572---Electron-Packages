```python
# Clean up data by mapping regex across our data
dataframe_just_dependencies = dataframe.drop(['Repository'], axis=1)
dataframe_just_dependencies = dataframe_just_dependencies.applymap(lambda x: x if pd.isna(x) else pd.Series(x).str.replace(r':\s(^|\d|>?).*', '', regex=True)[0])

dataframe_list = pd.DataFrame(dataframe['Repository'])
dataframe_list['Dependecies'] = dataframe_just_dependencies.values.tolist()
dataframe_list['Dependecies'] = dataframe_list['Dependecies'].apply(lambda x: [i for i in x if not pd.isna(i)])

display(dataframe_list)

dataframe_dev_just_dependencies = dataframe_dev.drop(['Repository'], axis=1)
dataframe_dev_just_dependencies = dataframe_dev_just_dependencies.applymap(lambda x: x if pd.isna(x) else pd.Series(x).str.replace(r':\s(^|\d|>?).*', '', regex=True)[0])

dataframe_dev_list = pd.DataFrame(dataframe_dev['Repository'])
dataframe_dev_list['Dependecies'] = dataframe_dev_just_dependencies.values.tolist()
dataframe_dev_list['Dependecies'] = dataframe_dev_list['Dependecies'].apply(lambda x: [i for i in x if not pd.isna(i)])

display(dataframe_dev_list)
```

```python
# Find the most common non-dev dependencies:
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)

def findMostCommon(dfOfInterest):
  mostCommon = pd.DataFrame({'Dependency':[], 'Count':[]})

  for i in range(len(dfOfInterest)):
    for dependency in dfOfInterest['Dependecies'].iloc[i]:
      # Increment count if it's already in there
      if dependency in mostCommon['Dependency'].values:
        mostCommon.loc[mostCommon['Dependency'].isin([dependency]), 'Count'] += 1

      else: # Append to df if value not in there alrady
        row = pd.Series([dependency, 1], index=mostCommon.columns)
        mostCommon = mostCommon.append(row, ignore_index=True)

  # Sort the values
  mostCommonDesc = mostCommon.sort_values(by=['Count'], ascending=False)
  display(mostCommonDesc)

  return mostCommonDesc
```

```python
# This cell takes 10 minutes =(
print("Most common non-dev dependencies")
nonDevMostCommon = findMostCommon(dataframe_list)

print("Most common dev dependencies")
devMostCommon = findMostCommon(dataframe_dev_list)
```

```python
# Create a list excluding the top 5% of dependencies for non dev dependencies
nonDevExclusion = nonDevMostCommon[nonDevMostCommon['Count'] > nonDevMostCommon['Count'].quantile(0.91)]
display(nonDevExclusion)
nonDevExclusionSet = set(nonDevExclusion['Dependency'].values)

# Ditto for dev dependencies
devExclusion = devMostCommon[devMostCommon['Count'] > devMostCommon['Count'].quantile(0.91)]
display(devExclusion)
devExclusionSet = set(devExclusion['Dependency'].values)
```

```python
import os.path


def generateGephiFiles(dfList, exclusionSet, fileName):
  maxDep = 0
  nodes = pd.DataFrame({'ID':[], 'label':[], 'Dependencies':[]})
  edges = pd.DataFrame({'Source':[], 'Target':[], 'Weight':[], 'label':[]})

  # Loop through all unique pairs in the dataframe, O(n^2)
  for i in range(len(dfList)):
      for j in range(i, len(dfList)):
        if i == j:
          continue

        # Get the unique dependencies between two different repositories
        similar = set(dfList['Dependecies'].iloc[i]).intersection(set(dfList['Dependecies'].iloc[j])) # See what this looks like and if we can concatenate it to the edge
        # Remove the excluded dependencies
        similar = similar - exclusionSet

        if (len(similar) > 0) :
          # Get the weight and then get the max for normalization after we are done the loop
          weight = len(similar)
          maxDep = max(weight, maxDep)

          # Write node (repository name), label, and dependencies to nodes. Avoid putting same node in twice.
          if dfList['Repository'].iloc[i] not in nodes['label'].values:
            rowNodeI = pd.Series([dfList['Repository'].iloc[i], dfList['Repository'].iloc[i], dfList['Dependecies'].iloc[i]], index=nodes.columns)
            nodes = nodes.append(rowNodeI, ignore_index=True)
          if dfList['Repository'].iloc[j] not in nodes['label'].values:
            rowNodeJ = pd.Series([dfList['Repository'].iloc[j], dfList['Repository'].iloc[j], dfList['Dependecies'].iloc[j]], index=nodes.columns)
            nodes = nodes.append(rowNodeJ, ignore_index=True)


          # Write source (repo i), target (repo j), weight, and label to edge list
          rowEdge = pd.Series([dfList['Repository'].iloc[i], dfList['Repository'].iloc[j], weight, str(similar)], index=edges.columns)
          edges = edges.append(rowEdge, ignore_index=True)

  # Normalize the weights inside the edge list and round to the fifth decimal
  edges['Weight'] = edges['Weight'].div(maxDep).round(5)

  print("Node list:")
  display(nodes)

  print("Edge list:")
  display(edges)

  # Now upload the node and edge lists to our data folder
  nodes.to_csv('/content/drive/MyDrive/CPSC 572 Project/Data/'+ fileName +'NodeList.csv', sep='\t', index=False)
  if os.path.isfile('/content/drive/MyDrive/CPSC 572 Project/Data/'+ fileName +'NodeList.csv'):
    print("Successfully uploaded "+ fileName +" node list to /content/drive/MyDrive/CPSC 572 Project/Data/")
  else:
    print("ERROR: Download unsuccessful for " + fileName + " to /content/drive/MyDrive/CPSC 572 Project/Data/")

  edges.to_csv('/content/drive/MyDrive/CPSC 572 Project/Data/'+ fileName +'EdgeList.csv', sep='\t', index=False)
  if os.path.isfile('/content/drive/MyDrive/CPSC 572 Project/Data/'+ fileName +'EdgeList.csv'):
    print("Successfully uploaded "+ fileName +" edge list to /content/drive/MyDrive/CPSC 572 Project/Data/")
  else:
    print("ERROR: Download unsuccessful for " + fileName + " to /content/drive/MyDrive/CPSC 572 Project/Data/")

```

```python
# Generate the Gephi files as node and edge lists. See instructions below for how to load them into Gephi
# WARNING: This cell takes a while to run (1h10min~)
  # There are some optimizations that can be made using "isin" instead of a nested loop to find matching values
  # There's also using concat instead of append, but that requires some transpose magic

generateGephiFiles(dataframe_list, nonDevExclusionSet, "nonDevDep")
generateGephiFiles(dataframe_dev_list, devExclusionSet, "devDep")
```
